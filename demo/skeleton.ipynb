```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mmcv import load, dump

annotations = load('../data/nturgbd/ntu60_3danno.pkl')['annotations']

# Example data: 25 joints with 3 coordinates each
num_joints = 25
num_dimensions = 3
# data = np.random.rand(num_joints, num_dimensions)
```

```python
data = annotations[1]['keypoint'][0][10]
```

```python
# Create a figure and a 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the joints in 3D
for i in range(num_joints):
    x, y, z = data[i, 0], data[i, 1], data[i, 2]
    ax.scatter(x, y, z, color='b')  # You can customize color and size if needed

# Set axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the aspect ratio to be equal
ax.set_box_aspect([np.ptp(data[:,0]), np.ptp(data[:,1]), np.ptp(data[:,2])])

plt.show()
```
